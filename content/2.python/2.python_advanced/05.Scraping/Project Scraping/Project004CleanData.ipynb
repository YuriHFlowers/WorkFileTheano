```python
import pandas as pd
import numpy as ny
```

```python
part1= pd.read_csv('./FINAL2.csv')
part2= pd.read_csv('./FINAL.csv')
postcode= pd.read_csv('./URLcode.csv')
```

```python
del part1['Unnamed: 0']
del part2['Unnamed: 0']
del part2['Price']
del postcode['Unnamed: 0']
```

```python
result = postcode.URL.str.extractall(pat = '([0-9]{4}\/)')
result = result.reset_index()
```

```python
del result['match']
del result['level_0']
result = result.set_axis(['Postal_Code'], axis=1)
```

```python
result = result.Postal_Code.str.extractall(pat = '([0-9]{4})')
```

```python
result = result.reset_index()
```

```python
del result['level_0']
del result['match']
result = result.set_axis(['Postal_Code'], axis=1)
```

```python
result.head()
```

```python
immoweb = pd.concat([result, part1, part2], axis=1, join="inner")
immoweb.head()
```

```python
def codingYesNo(x):
    if x =='No':
        return '0'
    if x =='Yes':
        return '1'

immoweb['Furnished_code'] = immoweb['Furnished'].apply(codingYesNo)
immoweb['Swimming_pool_code'] = immoweb['Swimming_pool'].apply(codingYesNo)
immoweb.head(2)
```

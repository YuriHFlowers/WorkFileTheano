### Bonus: Reindex to Align with Other Objects

```python
drinks.set_index('country', inplace=True)
drinks.info()
# I see that is an indexed dataFrame
```

```python
drinks.head()
```

```python
# Notice the df is indexed
# Select some variables of the df
object1 = drinks[['continent','beer_servings','spirit_servings']]
object1
```

```python
object2 = drinks.loc['Afghanistan':'Angola', ['continent','beer_servings','spirit_servings']]
object2
```

```python
object1 = object1.reindex_like(object2)
object1  
# I reindexed the object 1, in this example its exactly the same information. 
# but it is only an educational example. I will do another example.

# Note: Here, the df1 DataFrame is altered and reindexed like df2. 
# The column names should be matched or else NAN will be added for the entire column label.
```

### Bonus: Reindex to Align with Other Objects (other example)

```python
obj2 = drinks.iloc[0:20, 1:4]
obj2
```
